## Make the data workable

```python
import re, json

with open("task1_d.json", "r", encoding="utf-8") as f:
    text = f.read()
    
text = text.lstrip("\ufeff").strip()

text = re.sub(r':([A-Za-z0-9_]+)\s*=>', r'"\1":', text)

data = json.loads(text)

print("Just experiment:", data[0])
```

## Loading the json data to mysql database

```python
import re
import json
import mysql.connector

# I created a database through terminal and a table through Azure Data Studio, then just applied an insert through python

conn = mysql.connector.connect(
    host="localhost",
    user="Akbota",
    password="Akbota_2003",
    database="task1_db"
)
cursor = conn.cursor()
```

```python
sql = """
INSERT INTO books (id, title, author, genre, publisher, year, price)
VALUES (%s, %s, %s, %s, %s, %s, %s)
"""

for i, book in enumerate(data, start=1):
    cursor.execute(sql, (
        book["id"],
        book["title"],
        book["author"],
        book["genre"],
        book["publisher"],
        book["year"],
        book["price"]
    ))

    # I had some problems with uploading the data as the process took several hours, that's why I added the checkings when some data was added
    if i % 50 == 0:
        print(f"Inserted {i} rows...")

conn.commit()
cursor.close()
conn.close()

print("Upload completed successfully!")
```
